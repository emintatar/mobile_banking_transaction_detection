```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv("mobile_bank_dataset.csv")
```

```python
df = dataset.copy()
```

```python
df.head()
```

```python
df = df[["sorgu", "label"]]
```

```python
df.head()
```

```python
df["sorgu"].unique()
```

```python
df["label"].unique()
```

```python
stopwords = ['fakat','lakin','ancak','acaba', 'ama', 'aslında', 'az', 'bazı', 'belki', 'biri', 'birkaç', 'birşey', 'biz', 'bu', 'çok', 'çünkü', 'da', 'daha', 'de', 'defa', 'diye', 'eğer', 'en', 'gibi', 'hem', 'hep', 'hepsi', 'her', 'hiç', 'için', 'ile', 'ise', 'kez', 'ki', 'kim', 'mı', 'mu', 'mü', 'nasıl', 'ne', 'neden', 'nerde', 'nerede', 'nereye', 'niçin', 'niye', 'o', 'sanki', 'şey', 'siz', 'şu', 'tüm', 've', 'veya', 'ya', 'yani']
```

```python
for word in stopwords:
    word = " " + word + " "
    df['sorgu'] = df['sorgu'].str.replace(word, " ")
```

```python
df.head()
```

```python
cv = CountVectorizer(max_features = 50)
x = cv.fit_transform(df['sorgu']).toarray()
y = df['label']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 10, train_size = 0.7)
```

```python
rf = RandomForestClassifier()
model = rf.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
message = input("Yapmak istediginiz islemi giriniz: ")
message_df = pd.DataFrame({"sorgu": message ,"label": 0}, index = [20])
df = pd.concat([df ,message_df], ignore_index = True)

for word in stopwords:
    word = " " + word + " "
    df['sorgu'] = df['sorgu'].str.replace(word, " ")

cv = CountVectorizer(max_features = 50)

x = cv.fit_transform(df['sorgu']).toarray()
y = df['label']

guess = x[-1].copy()
x=x[0:-1]
y=y[0:-1]

x_train, x_test ,y_train, y_test = train_test_split(x, y, random_state = 11, train_size = 0.7)

rf = RandomForestClassifier()
model = rf.fit(x_train, y_train)
score = model.score(x_test, y_test)

result = model.predict([guess])

print("Sonuc:", result, "Skor:", score)
```
